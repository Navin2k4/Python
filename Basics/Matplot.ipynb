Bar Chart

```python
import matplotlib.pyplot as plt

a = [1, 2, 3]   # x values
b = [5, 8, 3]   # height values

plt.bar(a, b)
plt.show()
```

Scatter Plot

```python
x1 = np.array([1,2,4,5,6,5])
y1 = np.array([324,423,356,456,563,324])
plt.scatter(x1,y1)
plt.show()
```

Pie Chart

```python
p = np.array([4,5,2,4,5])
plt.pie(p)
plt.show()
```
